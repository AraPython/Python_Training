# **Pandas Series**
A Pandas Series is like a column in a tableIt is a one-dimensional array holding data of any type.


```
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```




````python
import pandas as pd

def max(myvar):
  a= []
  for i in myvar:
    a.append( float(i)) #عملية تحويل القيمة من داتا تايب ل داتا اخر, على سبيل التحديد هنا حولنا الة رقم كسري 
    #شرط التحويل لفلوت هو ان يكون الستريق عبارة عن فلوت
  return pd.Series(a).max()
a = [1, 7.0, "5555.55"]

myvar = pd.Series(a)
print(max(myvar))#دالة احنا كتبناها 
#print(myvar.min())#دالة من البانداس مباشرة

#45دقيقة 
````
