```python
class Car:

  """A simple attempt to represent a car."""
  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year
    self.__private_odometer_reading = 0
    self.color = "color"


  def get_descriptive_name(self):
    #long_name = f"{self.year} {self.make} {self.model}"
    long_name = str(self.year) + " " + self.make + " " + self.model + "Color is " + self.color
    """
    do somthing
    """
    return long_name.title()

    
  def read_odometer(self):
    #print("This car has "+self.odometer_reading+" miles on it.")
    print(f"This car has {self.__private_odometer_reading} miles on it.")

  def get_Color(self):
    #print(f"This car color id {self.color}")
    return self.color

  def set_Color(self, color="white"):
    self.color = color  
  
  def update_odometer(self, mileage):
    if mileage >= self.__private_odometer_reading:
      self.__private_odometer_reading = mileage
    else:
      print("You can't roll back an odometer!")


  def increment_odometer(self, miles):
    self.__private_odometer_reading += miles

class ElectricCar(Car):
  """Represent aspects of a car, specific to electric vehicles."""
  def __init__(self, make, model, year , color ):
    """Initialize attributes of the parent class."""
    super().__init__(make, model, year , color )
    
```

```python
car1 = Car(make="Honda",model="civic",year=2020 )
print(car1.get_descriptive_name())
```

```python
car1.read_odometer()
car1.update_odometer(50)
car1.read_odometer()

#هذا الكود ما اشتغل لانها برايفت
car1.odometer_reading=100
car1.read_odometer()

car1.update_odometer(100)
car1.read_odometer()

car1.update_odometer(50)
car1.read_odometer()
```
