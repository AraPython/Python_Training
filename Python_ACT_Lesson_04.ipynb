```python
for i in range(1,5):
  print(f" i = {i}")
  for j in range(1,3):
    #print(" j = {}".format(j)) # تابعة ثاني فور
    print(f"i={i}, j={j}, i*j= {i*j}") # تابعة ثاني فور
  print(f'end second loop iteration')# تابعة لاول فور
print('end first loop')# برا كل الفور
```

```python
a=[] # عملنا لستة فارغة
for i in range(5): # عملنا فور لووب من 0 الى 4
  a.append([i for i in range(5)]) # باستخدام الدالة append اضفنا 5 ليستات جديدة  

for row in a:# دالة طباعة سطر سطر
  print(row)

for row in a:# دالة طباعة عنصر عنصر, في كل سطر بنطبع كل العناصر
  for itm in row:
    print(itm)
```
