# **Python functions**
A function is a block of Python code that robotically does the same thing again
and again, whenever you invoke its name. It saves you repetitive coding and
makes your code easier to understand.
We'll start with a trivially simple example.


```
def add_numbers():
  first_number = 2
  second_number = 3
  total = first_number + second_number
  print(total)
```



````python
def add_numbers():
  n1=9
  n2=9
  total = n1+n2
  print(total)
````

````python
add_numbers()
add_numbers()
for i in range(5):
  add_numbers()
````

# **Dealing with an unknown number of arguments**


```

```



```
def display_result(winner, score, **other_info):
  print("The winner was " + winner)
  print("The score was " + score)
  for key, value in other_info.items():
  print(key + ": " + value)
```



````python
def display_result(winner="", score="", **other_info):
  if len(winner)>0:
    print(f"The winner was {winner}")
    print(f"The score was {score}")
  for key, value in other_info.items():
    print(f"{key}  {value}")
````

# **Local vs. global variables**
A global variable is one you define in the main body of your code—that is,
not in a function.
A local variable is one that you define in a function.

````python
a = 15
#a = [x for x in range(5)] 
#print(x)
def foo():
  a=5
  b=6
  print(a)
  return a+b
print(foo())
print(a)
def foo2():
  a=5
  b=6
  print(a)
  return a+b
````

````python
a=15
def foo():
  a=5
print(a)
````

````python
a=15
def foo():
  global a
  a=6
foo()
print(a)
````

# **Functions within functions**

````python
def fn1():
  return 5
def fn2():
  return 5
def fun_add_fun():
  return fn1()+fn2()
print(fun_add_fun())
````

````python
def factorial_recursion(n):
  if n>1:
    n = n * factorial_recursion(n-1)
  return n
````

````python
print(factorial_recursion(9))
````
