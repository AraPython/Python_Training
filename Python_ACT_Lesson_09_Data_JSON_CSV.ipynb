#**bold text** Parse JSON - Convert from JSON to Python


```
import json

# some JSON:
x =  '{ "name":"John", "age":30, "city":"New York"}'

# parse x:
y = json.loads(x)

# the result is a Python dictionary:
print(y["age"])
```



````python
import json

# some JSON:
x =  '{ "name":"Mahmoud", "age":36, "city":"Gaza"}'
# parse x:
y = json.loads(x)# هدفها تحول الجيسون الى ديكشيناري للتعامل مع البايثون
# the result is a Python dictionary:
print(y["age"])
````

# **Convert from Python to JSON**


```
import json

# a Python object (dict):
x = {
  "name": "John",
  "age": 30,
  "city": "New York"
}

# convert into JSON:
y = json.dumps(x)

# the result is a JSON string:
print(y)
```



````python
x = { # بايثون ديكشوناري يعتوي على مجموعة من المعلومات
  "name": "'محمود",
  "age": 36,
  "city": "غزة",
  2: "غزة" #D لاحظ كيف الانتجر تحول لسترينج خلال عملية التحويل  "2": "غزة"
}
# الدالة dumps 
# تقوم بتحويل البايثون ديكشوناري لجيسون
# ensure_ascii=False للتعامل مع الاحرف العربية
y = json.dumps(x, ensure_ascii=False,indent=4)
print(y)
#print(y["city"]) بتعطي ايررور لانها جسون و ليست ديكشوناري
print(type(y)) # لنشوف نوع المتغير 
````

# **Writing to a CSV file Using csv.writer**



```
import csv

header = ['Name', 'M1 Score', 'M2 Score']
data = [['Alex', 62, 80], ['Brad', 45, 56], ['Joey', 85, 98]]

filename = 'Students_Data.csv'
with open(filename, 'w', newline="") as file:
    csvwriter = csv.writer(file) # 2. create a csvwriter object
    csvwriter.writerow(header) # 4. write the header
    csvwriter.writerows(data) # 5. write the rest of the data
```



````python
import csv

header = ['Name', 'M1 Score', 'M2 Score']
data = [['Alex', 62, 80], ['Brad', 45, 56], ['Joey', 85, 98]]

filename = 'Students_Data.csv'
with open(filename, 'w', newline="") as file:
    csvwriter = csv.writer(file) # 2. create a csvwriter object
    csvwriter.writerow(header) # 4. write the header
    csvwriter.writerows(data) # 5. write the rest of the data
````

# **Writing** to a CSV file Using writelines()

```
header = ['Name', 'M1 Score', 'M2 Score']
data = [['Alex', 62, 80], ['Brad', 45, 56], ['Joey', 85, 98]]
filename = 'Student_scores.csv'
with open(filename, 'w') as file:
    for header in header:
        file.write(str(header)+', ')
    file.write('n')
    for row in data:
        for x in row:
            file.write(str(x)+', ')
        file.write('n')
```
